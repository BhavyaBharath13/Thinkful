```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("MonthlyAirlineInfoIndia.csv")
```

```python
df.sort_values(['FREIGHT FROM INDIA'],ascending = False).head(10)
```

```python
#Plot to display Frieght from India. Data visualization is for years 2015 to 2017. There is a sharp raise in Frieght in Dec 2015
#This was because Etihad Airways added six new Global Destinations and Kolkota, India was one of them.  
plt.figure(figsize=(20,10))
y = df['FREIGHT FROM INDIA']
plt.plot(y)
plt.xlim(0,2500)
plt.ylabel('Freight')
plt.title('Freight from India')
plt.show()
```

```python
#Represents Passengers from India who travelled in Domestic as well as International flights from 2015-2017.  
#The data represents monthly breakups. 
plt.figure(figsize = (20,10))
slice = df.iloc[:,[1,6]]
bp = slice.boxplot(column='PASSENGERS FROM INDIA', by='MONTH')
axes = plt.gca()
plt.xlabel('Month')
plt.ylabel('No Of Passengers')
plt.show()
```

```python
#Below demonstrates the sub plots for 4 different scenarios
#The 1st and 2nd plots provides information on Frights from(1st) and to(2nd) India for each quarter
#The 3rd and 4th plots provides information on Passengers from(3rd) and to(4th) India for each quarter

plt.figure(figsize=(20, 10))

plt.subplot(141)
plt.scatter(x = df['QUARTER'], y = df['FREIGHT FROM INDIA'], color='purple', marker='+', s=50)
plt.xlabel('Quarter')
plt.ylabel('Frieght')
plt.title('Quarter v/s Freight from India')

plt.subplot(142)
plt.scatter(x = df['QUARTER'], y = df['FREIGHT TO INDIA'], color='pink', marker='+', s=50)
plt.xlabel('Quarter')
plt.ylabel('Freight')
plt.title('Quarter v/s Freight to India')

plt.subplot(143)
plt.scatter(x = df['QUARTER'], y = df['PASSENGERS FROM INDIA'], color='green', marker='+', s=50)
plt.xlabel('Quarter')
plt.ylabel('Passengers')
plt.title('Quarter type v/s Passengers from India')

plt.subplot(144)
plt.scatter(x = df['QUARTER'], y = df['PASSENGERS TO INDIA'], color='blue', marker='+', s=50)
plt.xlabel('Quarter')
plt.ylabel('Passengers')
plt.title('Quarter type v/s Passengers to India')

plt.tight_layout()
plt.show()
```

```python
#Histogram to show the plot of Passengers to India v/s from India over 2015-2017
plt.figure(figsize=(20, 10))
#y = df['FREIGHT TO INDIA']
#x = df['FREIGHT FROM INDIA']
ymean = np.mean(df['FREIGHT TO INDIA'])
ystd = np.std(df['FREIGHT TO INDIA'])
xmean = np.mean(df['FREIGHT FROM INDIA'])
xstd = np.std(df['FREIGHT FROM INDIA'])
x = np.random.normal(xmean,xstd,3000)
y = np.random.normal(ymean,ystd,3000)
plt.hist(x, density = True, color=['b'], bins='auto', alpha=.5) 
plt.hist(y, density = True, color=['r'], bins='auto', alpha=.5)
plt.show()
```
