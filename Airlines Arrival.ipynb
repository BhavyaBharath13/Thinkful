```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#importing data
data = pd.read_csv("2000.csv")
```

```python
data.shape
```

```python
#Too many rows, lets see the columns
data.head()
```

```python
data.describe()
```

```python
#Dropping unecessary columns
data.drop(columns = "CancellationCode", inplace = True)
data.drop(columns = "CarrierDelay", inplace = True)
data.drop(columns = "WeatherDelay", inplace = True)
data.drop(columns = "NASDelay", inplace = True)
data.drop(columns = "SecurityDelay", inplace = True)
data.drop(columns = "LateAircraftDelay", inplace = True)
data.drop(columns = 'Year', inplace = True)
data.drop(columns = 'Cancelled', inplace = True)
data.drop(columns = 'Diverted', inplace = True)
```

```python
#Take sample of 10% of data would still be more than 100,000
sample = data.sample(frac = 0.10)
```

```python
#Get rid of all the rows with null values
sample.dropna(inplace = True)
```

```python
sample
```

```python
#Feature Engineering
sample['Late'] = np.where(sample.ArrDelay > 30,1,0)
sample['Arrival'] = sample.ArrTime - sample.CRSArrTime
sample['Departure'] = sample.DepTime - sample.CRSDepTime
sample['DepArr'] = sample.Departure - sample.Arrival
sample['Scheduled'] = sample.CRSArrTime + sample.CRSDepTime
sample['Taxi'] = sample.TaxiIn + sample.TaxiOut
sample['Speed'] = sample.Distance / sample.AirTime
```

```python
#Features for total delay
sample['Delay'] = sample.ArrDelay + sample.DepDelay
```

```python
sns.countplot(x= 'Late',data = sample)
```

```python
#Not that biased
#Lets see correlations
plt.figure(figsize=(15, 10))
print(sns.heatmap(sample.corr()))
plt.show()
sample.corr()
```

```python
from sklearn.linear_model import LogisticRegression

#Dont want to take ArrDelay or DepDelay so using features Delay and Taxi
X = sample[['DepTime','Delay','Taxi']]
y = sample[['Late']]
offset = int(X.shape[0] * 0.25)

# Put 75% of the data in the training set.
X_train, y_train = X[:offset], y[:offset]

# And put 25% in the test set.
X_test, y_test = X[offset:], y[offset:]

lr = LogisticRegression(C=1)
lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(lr, X_train, y_train, cv=5)
```

## Conlusion:

Since the data was a lot, I have used only 10% of data as my sample of interest and the Delay is calculated using the ArrDelay column. I know that we can use the difference between Scheduled Arrival time and actual arrival time for this as well but I have used the given column instead. I feel that this is one of the examples where feature engineering works very well. We got a very good score of 98% accuracy overall, with good computing power I would like try predicting for the entire data in future.
