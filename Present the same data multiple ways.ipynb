```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
titanic = sns.load_dataset("titanic")
titanic.head(5)
```

```python
sns.set(style="ticks")
```

```python
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.suptitle('Passenger Class vs Count')

plt.subplot(2,2,1)
ax=sns.distplot(titanic.pclass, kde = False)
plt.title("Plot1")
plt.xlabel("PClass")
plt.ylabel("count")

plt.subplot(2,2,2)
plt.title("Plot2")
ay=sns.countplot(x="pclass",data = titanic,palette = 'Greens_d')

plt.subplot(2,2,3)
plt.title("Plot3")
bx = sns.boxplot(y="pclass",data = titanic,palette ='bright')
plt.xlabel("count")

plt.subplot(2,2,4)
plt.title("Plot4")
by = sns.pointplot(y="pclass",data = titanic,palette ='Reds_d')
plt.xlabel("count")

plt.show()
```

1. We see that in Plot 1, though there is a clear identification of Passenger Class(PClass), the x value shows as range between whole number to nearest point (Ex 1 to 1.25) which is incorrect.<br> 
2. In plot 2, the values make more sense in figuring out the no of people in each class to approximate value<br>
3. In plot 3, we are just representing the pclass value in terms of lowest, highest and mid. However this does not make any sense as we dont know how many people are there for each pclass and it is incorrect to say that there is a range between lowest, mid and highest values.<br>
4. Plot 4 just gives the mean value of pclass which is not useful in anyway. Here again there is not count of people.

```python
sns.set_style("whitegrid") 
plt.suptitle('Survived vs Age')

plt.subplot(1,3,1)
ax = sns.stripplot(x='survived', y='age', data=titanic, palette='dark')
plt.title("Plot5")

plt.subplot(1,3,2)
ax = sns.swarmplot(x='survived', y='age', data=titanic, palette='pastel')
plt.title("Plot6")

plt.subplot(1,3,3)
ax = sns.pointplot(x='survived', y='age', data=titanic, palette='muted')
plt.title("Plot7")

plt.show()
```

1. In Plot5, we can see that the values range from just after 0 to almost 80, however we dont exactly know the mean value of the age of people who survived.<br>
2. In Plot6, we see that though the age ranges from 0 to 80, we can see the most of the range is concentrated bewteen 20 and 40 for both who survived and did not.<br>
3. Plot7 exactly tells us the mean range and the mid values, however we don't get the idea of upper limit and lower limit distributions for age.

```python
sns.set_style("whitegrid") 
plt.suptitle('Who vs Age')

plt.subplot(2,3,1)
ax = sns.stripplot(x='who', y='age',hue='survived', data=titanic, palette='dark')
plt.title("Plot8")

plt.subplot(2,3,2)
ax = sns.swarmplot(x='who', y='age',hue='survived', data=titanic, palette='pastel')
plt.title("Plot9")

plt.subplot(2,3,3)
ax = sns.barplot(x='who', y='age',hue='survived', data=titanic, palette='muted')
plt.title("Plot10")

plt.subplot(2,3,4)
ax = sns.boxenplot(x='survived', y='age',hue='who', data=titanic, palette='bright')
plt.title("Plot11")

plt.subplot(2,3,5)
ax = sns.violinplot(x='survived', y='age',hue='who', data=titanic, palette='bright')
plt.title("Plot12")

plt.subplot(2,3,6)
ax = sns.pointplot(x='survived', y='age',hue='who', data=titanic, palette='colorblind')
plt.title("Plot13")

plt.show()
```

1. Plots 8 through 10 have data in terms of who survived and who didn't based on ages. Though Plots 8 and 10 are similar looking, plot 8 has right value range for adults and children, where as plot 10 looks incorrect as range for man and woman(adults) start at 0. Plot 9 (swarm plot) actualy tells the real story for this type of plotting giving info on range as well as concentration of each type for ages.<br>
2. Plots 11 through 13 have data in terms of survival and age based on kind of survivor(man, women or child). Here, plots 11 and 12 look similar but differentiating the means for plot 12 looks a bit difficult for comparisons where as in plot11 there is a clear comparison of means and highest and lowest values. Plot 13, though tells us the mean age range doesn't expalin this scenario on its own.
