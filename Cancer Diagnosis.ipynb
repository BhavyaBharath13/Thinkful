```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#importing data
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
#Classifying data
data['Diag'] = np.where(data.diagnosis == 'M',1,0)
```

```python
#Lets see how our data looks
sns.countplot(x='Diag',data=data, palette = 'cubehelix')
```

```python
#Not completely imbalanced, dont need the object field since we classified it
data.drop(columns='diagnosis',inplace=True)
```

```python
data.describe()
```

```python
#Dropping Unnamed as data cleanup process
data.drop(columns = 'Unnamed: 32',inplace = True)
```

```python
# Make the correlation matrix.
corrmat = data.corr()
print(corrmat)

# Set up the matplotlib figure.
f, ax = plt.subplots(figsize = (12,9))

# Draw the heatmap using seaborn.
sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()
```

```python
#Finding largest n correlation greater than 0.5
np.absolute(data[data.columns[1:]].corr()['Diag'][:-1]).sort_values(ascending = False)
```

```python
#Using KNN model 1st without weights
from sklearn import neighbors

# Build our model.
knn = neighbors.KNeighborsRegressor(n_neighbors=10)
X = data[['concave points_worst',
'perimeter_worst',
'concave points_mean',
'radius_worst',
'perimeter_mean',
'area_worst',
'radius_mean',
'area_mean',
'concavity_mean',
'concavity_worst',
'compactness_mean',
'compactness_worst',
'radius_se',
'perimeter_se',
'area_se'
]]
Y = data.Diag
knn.fit(X, Y)
```

```python
#Score for over all data
knn.score(X,Y)
```

```python
X.describe()
```

```python
#Lets scale and again try fitting
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
ft = scaler.fit(X).transform(X)
```

```python
ft.max()
```

```python
knn.fit(ft,Y)
```

```python
knn.score(ft,Y)
```

```python
#Good score2 lets split to train and test now
ft.shape
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(ft, Y, test_size=0.3, random_state=0)
```

```python
knn.fit(X_train,y_train)
```

```python
#Score for Training predictors 
knn.score(X_train,y_train)
```

```python
#Predict test set target
y_test = knn.predict(X_test)
```

```python
#Cross validate the scores

score = cross_val_score(knn, X, Y, cv=5)
print("Unweighted Accuracy whole: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
score = cross_val_score(knn, ft, Y, cv=5)
print("Unweighted Accuracy scaled: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
scoretr = cross_val_score(knn, X_train, y_train, cv=5)
print("Unweighted Accuracy scaled train: %0.2f (+/- %0.2f)" % (scoretr.mean(), scoretr.std() * 2))
scorete = cross_val_score(knn, X_test, y_test, cv=5)
print("Unweighted Accuracy scaled test: %0.2f (+/- %0.2f)" % (scorete.mean(), scorete.std() * 2))
```

```python
#LEts try weighted model by distance
knnw = neighbors.KNeighborsRegressor(n_neighbors=10, weights='distance')
knnw.fit(X,Y)
```

```python
#Try score on original data
knnw.score(X,Y)
```

```python
#Overfit???? Try on scaled data
knnw.fit(ft,Y)
```

```python
#Lets see scaled data score
knnw.score(ft,Y)
```

```python
# Oops! Is it overfitted ?? Try on train and test
knnw.score(X_train,y_train)
```

```python
y_test = knnw.predict(X_test)
```

```python
#Lets see Cross validation scores since it looks like weighted model is overfitting
score = cross_val_score(knnw, X, Y, cv=5)
print("Unweighted Accuracy whole: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
score = cross_val_score(knnw, ft, Y, cv=5)
print("Unweighted Accuracy scaled: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
scoretr = cross_val_score(knnw, X_train, y_train, cv=5)
print("Unweighted Accuracy scaled train: %0.2f (+/- %0.2f)" % (scoretr.mean(), scoretr.std() * 2))
scorete = cross_val_score(knnw, X_test, y_test, cv=5)
print("Unweighted Accuracy scaled test: %0.2f (+/- %0.2f)" % (scorete.mean(), scorete.std() * 2))
```

## Conclusion:

Looks like in this situation our KNN unweighted model worked better as the weighted on tried to overfit giving a score of 100%. Overall we got a realistic kind of score for the KNN unweighted model which was scaled and was split.
