```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#importing data
data = pd.read_csv("MELBOURNE_HOUSE_PRICES_LESS.csv")
```

```python
data.head()
```

```python
data.drop(columns = ['Address','SellerG'], axis = 1)
```

```python
data.describe()
```

```python
data['Price'].fillna(data['Price'].mean(),inplace=True)
```

```python
sns.pairplot(data)
```

```python
X= data[['Distance', 'Rooms', 'Propertycount']]
y = data['Price']
```

```python
from sklearn import linear_model
import warnings
warnings.filterwarnings(action="ignore", module="scipy", message="^internal gelsd")

# Instantiate our model.
regr = linear_model.LinearRegression()

# Fit our model to our data.
regr.fit(X, y)

# Display the attributes we calculated.
print('Coefficients: \n', regr.coef_)
print('Intercept: \n', regr.intercept_)
```

```python
regr.score(X,y)
```

```python
df = pd.get_dummies(data[['Suburb','Regionname','CouncilArea','Rooms','Type','Method']])
```

```python
result = pd.concat([data, df], axis=1, sort=False)
```

```python
result.drop(columns = ['Suburb','Address','Rooms','Type','Method','SellerG','Date','Regionname','CouncilArea'],inplace = True)
```

```python
result
```

```python
result.dropna(inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

y = result.Price.values.reshape(-1, 1)
X = result.loc[:, ~(result.columns).isin(['Price'])]
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
regr.score(X, y)
```

```python
cross_val_score(regr, X, y, cv=5)
```

```python

from sklearn.cross_decomposition import PLSRegression
# Save predicted values.
Y_pred = regr.predict(X)
print('R-squared regression:', regr.score(X, y))

# Fit a linear model using Partial Least Squares Regression.
# Reduce feature space to 3 dimensions.
pls1 = PLSRegression(n_components=30)

# Reduce X to R(X) and regress on y.
pls1.fit(X, y)

# Save predicted values.
Y_PLS_pred = pls1.predict(X)
print('R-squared PLSR:', pls1.score(X, y))

# Compare the predictions of the two models
plt.scatter(Y_pred,Y_PLS_pred) 
plt.xlabel('Predicted by original 438 features')
plt.ylabel('Predicted by 30 features')
plt.title('Comparing LR and PLSR predictions')
plt.show()
```

```python
from sklearn.cross_decomposition import PLSRegression
# Save predicted values.
Y_pred = regr.predict(X)
print('R-squared regression:', regr.score(X, y))

# Fit a linear model using Partial Least Squares Regression.
# Reduce feature space to 3 dimensions.
pls1 = PLSRegression(n_components=10)

# Reduce X to R(X) and regress on y.
pls1.fit(X, y)

# Save predicted values.
Y_PLS_pred = pls1.predict(X)
print('R-squared PLSR:', pls1.score(X, y))

# Compare the predictions of the two models
plt.scatter(Y_pred,Y_PLS_pred) 
plt.xlabel('Predicted by original 438 features')
plt.ylabel('Predicted by 10 features')
plt.title('Comparing LR and PLSR predictions')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=500)
rfr.fit(X,y)
```

```python
rfr.score(X,y)
```

```python
importances = pd.DataFrame(index=X.columns)
importances['Importance'] = rfr.feature_importances_
importances.sort_values('Importance', ascending=False)
```

```python
price_fn = pd.DataFrame(index=X.columns)
price_fn['Coefficients'] = pls1.coef_
price_fn.sort_values('Coefficients', ascending=False)
```

```python
from sklearn import metrics
y_pred = pls1.predict(X)
print('RMSE:',np.sqrt(metrics.mean_squared_error(y,y_pred)))
```
